```python
import pandas as pd
df=pd.read_csv("StudentsPerformance.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df.groupby('gender')[['math score', 'reading score', 'writing score']].mean()
df.groupby('lunch')[['math score', 'reading score', 'writing score']].mean()
df.groupby('test preparation course')[['math score', 'reading score', 'writing score']].mean()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x='gender', y='math score', data=df)
plt.title("Math Score by Gender")
plt.xlabel("Gender")
plt.ylabel("Math Score")
plt.show()


sns.histplot(df['reading score'], kde=True)
plt.title("Distribution of Reading Scores")
plt.xlabel("Reading Score")
plt.ylabel("Frequency")
plt.show()


sns.barplot(x='lunch', y='writing score', data=df)
plt.title("Average Writing Score by Lunch Type")
plt.xlabel("Lunch Type")
plt.ylabel("Writing Score")
plt.show()


sns.barplot(x='test preparation course', y='reading score', data=df)
plt.title("Average Reading Score by Test Preparation")
plt.xlabel("Test Preparation Status")
plt.ylabel("Reading Score")
plt.show()


sns.histplot(df['math score'], kde=True, color='skyblue')
plt.title("Distribution of Math Scores")
plt.xlabel("Math Score")
plt.ylabel("Frequency")
plt.show()
```

### 📌 Insights

- **Female students excel in reading and writing:**  
  On average, girls achieved reading scores of about **72** and writing scores around **74**. In contrast, boys scored lower, averaging approximately **65** in reading and **63** in writing.

- **Male students lead slightly in math:**  
  The average math score for male students was around **69**, which is a bit higher than the **65** scored by female students.

- **Lunch type affects performance:**  
  Students who received a **standard lunch** scored near **70** across all subjects. Those with **free or reduced lunch** scored noticeably lower — often **10 to 12 points** less.

- **Test preparation makes a difference:**  
  Students who **completed the test preparation course** had significantly better results, especially in reading and writing where scores reached around **74**, compared to lower averages for those who skipped the course.

- **Parental education influences scores:**  
  Higher student scores (**above 70**) were observed among those whose parents had a **bachelor’s or master’s degree**, while students with parents who only completed high school averaged around **63–65**.
